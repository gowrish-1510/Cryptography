```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
p = 0xFFFFFFFFFFFFFFFFFFFFFFFFFFFFFFFFFFFFFFFFFFFFFFFFFFFFFFFEFFFFFC2F
a = 0
b = 7

# Generate a range of x values
x = np.linspace(0, 2 **16, 10000)  

# Calculate y^2 = x^3 + a*x + b mod p for the actual curve equation
## p being a large integer poses a constraint here 
y_squared = (x**3 + a*x + b)

# Calculate both positive and negative y values
y_positive = np.sqrt(y_squared)
y_negative = -np.sqrt(y_squared)

# Plot the elliptic curve
plt.figure(figsize=(10, 6))
plt.plot(x, y_positive, 'r', linewidth=2, label='y = +sqrt(x^3 + 7)')
plt.plot(x, y_negative, 'b', linewidth=2, label='y = -sqrt(x^3 + 7)')
plt.title('SECP256K1 Elliptic Curve of Bitcoin')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```
